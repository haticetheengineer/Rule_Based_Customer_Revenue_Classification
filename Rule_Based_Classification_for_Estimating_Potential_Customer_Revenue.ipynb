**Task 2: What are the average earnings in the COUNTRY, SOURCE, SEX, and AGE breakdown?**


```python
df.groupby(['COUNTRY','SOURCE', 'SEX', 'AGE']).agg({'PRICE':'mean'})
```

**Task 3: Sort the output by PRICE**


```python
agg_df = df.groupby(['COUNTRY','SOURCE', 'SEX', 'AGE']).agg({'PRICE':'mean'}).sort_values('PRICE', ascending = False)
agg_df.head()
```

**Task 4: Convert the index names into variable names**

* Reset the index to convert the index columns into regular columns

```python
agg_df.reset_index(inplace = True)
agg_df
```

**Task 5: Convert the 'Age' variable into a categorical variable and add it to agg_df**

```python
#. Görev 5: Age değişkenini kategorik değişkene çeviriniz ve agg_df’e ekleyiniz.
agg_df['AGE'] = agg_df['AGE'].astype('category')
agg_df.info()
```

**Define age bins and labels for categorization**

```python
df['AGE_CAT'] = pd.cut(df['AGE'], bins = [0, 18, 23, 30, 40, 70],
                       labels = ('0_18', '19_23', '24_30', '31_40', '41_70'))
```

```python
agg_df['AGE_CAT'] = pd.cut(agg_df['AGE'], bins = [0, 18, 23, 30, 40, 70],
                       labels = ('0_18', '19_23', '24_30', '31_40', '41_70'))
```

```python
agg_df['AGE_CAT']
```

```python
agg_df
```

**Task 6: Define new tier-based customers (persona)**

```python
agg_df['customers_level_based'] = agg_df['COUNTRY'] + '_' + agg_df['SOURCE'] + '_' + agg_df['SEX'] + '_' + agg_df['AGE_CAT'].astype(str)
agg_df.head()
```

```python
agg_df.groupby('customers_level_based').agg({'PRICE':'mean'})
```

**Task 7: Segment the new customers (persona).**

* Divide the new customers (e.g., USA_ANDROID_MALE_0_18) into 4 segments based on PRICE.
* Add these segments to the agg_df DataFrame as a new variable named SEGMENT.
* Describe the segments (perform a group by on the segments and calculate the mean, max, and sum of PRICE).


```python
agg_df['SEGMENT'] = pd.qcut(agg_df['PRICE'], 4, labels = ['D', 'C', 'B', 'A'])
```

```python
agg_df
```

```python
agg_df.groupby('SEGMENT').agg({'PRICE':['mean', 'max', 'sum']})
```
